Import CSV from URL

```python
import pandas as pd
url = 'https://github.com/humphrjk/Business_Analytics/files/9398066/weather_data.csv'
df = pd.read_csv(url)
df
```

Dictionary

```python
weather_data ={
    'day':['1/1/2017','1/2/2017','1/3/2017'],
    'temperature':[32,35,28],
    'windspeed':[6,7,8],
    'event':['Ranin','Sunny','Snow']
}
df = pd.DataFrame(weather_data)
```

```python
df
```

Tuple

```python
weather_data =[
    ('1/12017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,8,'Snow')
]
df = pd.DataFrame(weather_data,columns=["day","temperature","windspeed","event"])
df
```

List of Dictionaries

```python
weather_data =[
                {'day':'1/1/2017','temperature':32,'windspeed':6,'event':'Rain'},
                {'day':'1/2/2017','temperature':35,'windspeed':7,'event':'Sunny'},
                {'day':'1/3/2017','temperature':28,'windspeed':8,'event':'Snow'}
               ]
df = pd.DataFrame(weather_data)
df
```
