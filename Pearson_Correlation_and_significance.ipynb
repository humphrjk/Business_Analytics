Load Pandas and import CSV file from URL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
url= 'https://github.com/humphrjk/Business_Analytics/files/9411284/diabetes_binary_5050split_health_indicators_BRFSS2015.csv'
df = pd.read_csv(url)

df.head()
```

```python
newdf = df[['Diabetes_binary','HighBP','HighChol','CholCheck','BMI','Smoker','HeartDiseaseorAttack','PhysActivity','Fruits','Veggies','HvyAlcoholConsump','AnyHealthcare','NoDocbcCost','GenHlth','MentHlth','PhysHlth','DiffWalk','Sex','Age','Education','Income']].dropna()
```

```python
pearsoncorr = df.corr(method='pearson').round(3)
pearsoncorr
```
