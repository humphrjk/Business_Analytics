Load Pandas and import CSV file from URL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
url= 'https://github.com/humphrjk/Business_Analytics/files/9418256/Shrewsbury.Survey.Fall.22.csv'
df = pd.read_csv(url)

df.head()
```

```python
pearsoncorr = df.corr(method='pearson').round(3)
pearsoncorr
```

```python
df.describe()
```

```python
sb.histplot(x=df['Q7'])
plt.gcf().set_size_inches(30,16)
plt.title("Q7 Distribution").fontsize=(20)
```
