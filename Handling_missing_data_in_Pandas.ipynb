**Handling missing data in pandas**
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html

```python
import pandas as pd
url = 'https://github.com/humphrjk/Business_Analytics/files/9440989/weather_data.csv'
#the day column is simply a string in a csv file. We need to change to a date type.
# to find out what the type of a column is use the function type(df.'day'[0])
df = pd.read_csv(url, parse_dates=["day"])
#Set the day as the index
#inplace euql to true to modify the dataframe otherwise it creates a new one
df.set_index('day',inplace=True)
df
```

***Fillna to change NaN to a value ***

```python
#created a new dataframe
new_df = df.fillna(0)
#fills with a zero and the use the new dataframe name to list the values
new_df
```

**Set a dictionary to handle changes in each column**

```python
#each columns can have a unique entry for NaN
newnew_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'

})
newnew_df
```

**Forward Fill method to take the previous value and copy into the NaN values as an estimate so that calucaltions like a MEAN aren't way off**

```python
df
```

```python
new3_df = df.fillna(method="ffill")
# you casn use bfill method to copy backward
new3_df
```

**Using Axis to copy horizintally **

```python
df
```

```python
new3_df = df.fillna(method="bfill", axis='columns')
# you casn use bfill method to copy backward
new3_df
```

```python
df
```

**Use limit to control how many times in  row the fill method can work**

```python
new3_df = df.fillna(method="ffill", limit=1)
# you casn use bfill method to copy backward
new3_df
```

**Interpolate This will come up with a linear value to insert into a cell between the values of the boundries of the empty cells**
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html

```python
new4_df = df.interpolate()
new4_df
```

**Use the "time" method to resolve the empty values in closer approximation to the closest time values ona linear or other curve**

```python
new5_df = df.interpolate(method="time")
new5_df
```

**Dropna function  drops any row with NaN values**

```python
new6_df = df.dropna()
new6_df
```

**Drop NA for only rows with all NaN**

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html

```python
new7_df = df.dropna(how="all")
new7_df
```

**Dropna with threshold keeps the row if it has atleast the number of the threshold value**

```python
new8_df = df.dropna(thresh=1)
new8_df
```

```python
new8_df = df.dropna(thresh=2)
new8_df
```

**Insert Date**
Create a date range and index
replace the index

```python
#creates the varible dt to define a range of dates
dt=pd.date_range("01-01-2017","01-11-2017")
#creates a varible of a new index of dates using dt
idx = pd.DatetimeIndex(dt)
#reindexes the dataframe with the idx
new9_df=df.reindex(idx)
new9_df
```
